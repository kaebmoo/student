**กำหนดเกณฑ์ส่วนสูง**

```python
def height_category (Ht):
  if Ht >= data_ht.iloc[0,0] : #Tall
    return data_ht.iloc[0,1]
  elif Ht >= data_ht.iloc[1,0] : #Medium
    return data_ht.iloc[1,1]
  elif Ht >= data_ht.iloc[2,0] : #Short
    return data_ht.iloc[2,1]
  else :
    return "No Data"

result['Ht Categories'] = result['Ht'].apply(height_category)

result.to_csv('apply_result.csv', index = False)
```
